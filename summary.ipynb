```python
from app.storage import load_data

def show_summary():
    data = load_data()

    income = data["income"]
    expenses = data["rent"] + data["food"] + data["utilities"] + data["leisure"]
    savings = income - expenses

    print("\n--- Résumé financier ---")
    print(f"Revenus : {income} €")
    print(f"Dépenses totales : {expenses} €")
    print(f"Économies : {savings} €")
```
