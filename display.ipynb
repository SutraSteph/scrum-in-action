```python
# affichage financier

# income revenu mensuel
# rent loyer mensuel
# food nourriture mensuelle
# utilities factures mensuelles
# leisure loisirs mensuels

# bugdet_rent
# budget_food
# budget_utilities
# budget_leisure

# total_budget
# total_expenses
# balance

income = 2500.00
rent = 800.00
food = 400.00
utilities = 150.00
leisure = 200.00    

budget_rent = 900.00
budget_food = 450.00
budget_utilities = 200.00
budget_leisure = 250.00

print("🎯 Catégories de dépenses")
print(f"🏠 Loyer       : {rent:.2f} €")
print(f"🍽️ Nourriture  : {food:.2f} €")
print(f"💡 Factures    : {utilities:.2f} €")
print(f"🎮 Loisirs     : {leisure:.2f} €")
print()

print("📊 Budget par catégorie")
print(f"🏠 Loyer       : {budget_rent:.2f} €")
print(f"🍽️ Nourriture  : {budget_food:.2f} €")
print(f"💡 Factures    : {budget_utilities:.2f} €")
print(f"🎮 Loisirs     : {budget_leisure:.2f} €")
print()

total_expenses = rent + food + utilities + leisure
balance = income - total_expenses

print("💰 Résumé financier")
print(f"💵 Revenu mensuel   : {income:.2f} €")
print(f"🧾 Dépenses totales : {total_expenses:.2f} €")
print(f"📌 Solde restant    : {balance:.2f} €")
```
