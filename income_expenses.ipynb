```python
import json
import os

FILE_PATH = os.path.join("data", "finances.json")


def add_income():
    # Demander le mois
    month = input("Entrez le mois (ex: 2026-03) : ")

    # Demander le montant du revenu
    amount = float(input("Entrez votre revenu mensuel : "))

    # Charger les données existantes ou créer la structure sinon
    if os.path.exists(FILE_PATH):
        with open(FILE_PATH, "r") as file:
            data = json.load(file)
    else:
        data = {
            "revenues": [],
            "expenses": []
        }

    # Vérifier que la clé revenues existe
    if "revenues" not in data:
        data["revenues"] = []

    # Créer l'objet revenu mensuel
    income = {
        "month": month,
        "amount": amount
    }

    # Ajouter à la liste des revenus
    data["revenues"].append(income)

    # Sauvegarder dans le fichier JSON
    with open(FILE_PATH, "w") as file:
        json.dump(data, file, indent=4)

    print("Revenu mensuel ajouté avec succès ")
```

```python
import json
import os

FILE_PATH = os.path.join("data", "finances.json")

ALLOWED_CATEGORIES = ["rent", "food", "leisure", "utilities"]

def add_expense():
    month = input("Entrez le mois (ex: 2026-03) : ")
    amount = float(input("Entrez le montant de la dépense : "))

    # indiquer la catégorie parmis celles indiqués . Sinon imprime catégorie invalide
    while True:
        category = input("Catégorie (rent, food, leisure, utilities) : ").lower()
        if category in ALLOWED_CATEGORIES:
            break
        print(" Catégorie invalide.")

    if os.path.exists(FILE_PATH):
        with open(FILE_PATH, "r") as file:
            data = json.load(file)
    else:
        data = {"revenues": [], "expenses": []}
    # evite erreur si expenses pas dans data => créer col expenses
    if "expenses" not in data:
        data["expenses"] = []
    #  créer Dict expenses :month - amount // Clé - valeur 
    expense = {"month": month, "amount": amount, "category": category}
    data["expenses"].append(expense)

    with open(FILE_PATH, "w") as file:
        json.dump(data, file, indent=4)

    print("Dépense mensuelle ajoutée avec succès ")
```
